```python
px_0 = 8
py_0 = 0
pz_0 = 20
step_n = 1500
v_0 = 0
a_0 = 0
j_0 = 0
K = 20
dt = 0.2
log = [[0,px_0,v_0,a_0,j_0]]
w1 = 3
final_time = 0.2 * step_n
w2 = 1
w3 = 1
w4 = 70
w5 = 40

# %% dimension
# % J 20*1
# % T 20*20
# %target_p 1 * 20
# %Bp 20 * 1
count = 1
```

```python
def getPredictionMatrix(K, dt, p_0, v_0, a_0):
    Tp = np.zeros([K, K])
    Tv = np.zeros([K, K])
    Ta = np.zeros([K, K])
    
    for i in range(K):
        Ta[i,:i] = np.ones([1,i])*dt
        for j in range(i):
            Tv[i,j] = (i-j+0.5)*dt**2
            Tp[i,j] = ((i-j+1)*(i-j)/2+1/6)*dt**3
        
    Ba = np.ones([K,1])*a_0
    Bv = np.ones([K,1])*v_0
    Bp = np.ones([K,1])*p_0
    
    for i in range(K):
        Bv[i] = Bv[i] + i*dt*a_0
        Bp[i] = Bp[i] + i*dt*v_0 + i**2/2*a_0*dt**2
    
    return Tp, Tv, Ta, Bp, Bv, Ba
```

```python
Tp, Tv, Ta, Bp, Bv, Ba = getPredictionMatrix(K, dt, px_0, v_0, a_0)
```

```python
up_v_xy = 6
bott_v_xy = 6
up_v_z = 6
bott_v_z = 1

up_a_xy = 3
bott_a_xy = 3
up_a_z = 3
bott_a_z = 1

up_j_xy = 3
bott_j_xy = 3
up_j_z = 2
bott_j_z = 2
```

```python
count = 0
Target_p_20 = np.asarray(x[count:count+K]).reshape(1,-1)
```

```python
H = scipy.linalg.block_diag(w4*np.ones([K,K]) + w1*(Tp.transpose()*Tp), 
                            w5*np.ones([K, K]))
```

```python
F = np.concatenate([2*w1*(np.matmul(Bp.transpose(),Tp)-np.matmul(Target_p_20,Tp)), 
            np.zeros([1,K])],axis=1)
```

```python
conc_with_identity = lambda x, y: np.concatenate([x, y*np.ones([K,K])],axis=1)
A = np.concatenate([conc_with_identity(Tv, -1),
                    conc_with_identity(-Tv, -1),
                    conc_with_identity(Ta, -1),
                    conc_with_identity(-Ta, -1),
                    conc_with_identity(np.ones([K,K]), -1),
                    conc_with_identity(-np.ones([K,K]), -1),
                    conc_with_identity(np.zeros_like(Ta), -1)],axis=0)
```

```python
b = np.concatenate([np.ones([K,1])*up_v_xy-Bv,
                    np.ones([K,1])*bott_v_xy+Bv,
                    np.ones([K,1])*up_a_xy-Ba,
                    np.ones([K,1])*bott_a_xy+Ba,
                    np.ones([K,1])*up_j_xy,
                    np.ones([K,1])*bott_j_xy,
                    np.zeros([K,1])],axis=0)
```

```python
import numpy as np
import scipy.sparse as sparse
import osqp
```

```python
P = sparse.csc_matrix(H)
q = np.array(F.T)
A_ = sparse.csc_matrix(A)
l = np.array(-b)
# l = None
u = np.array(b)

# Create an OSQP object
prob = osqp.OSQP()

# Setup workspace and change alpha parameter
prob.setup(P, q, A_, l, u, alpha=1.0)

# Solve problem
res = prob.solve()
```
